Importing the dependencies


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

Data Collection and Processing

```python
#loading dataset from sklearn
dataset = sklearn.datasets.load_breast_cancer()
```

```python
data_frame = pd.DataFrame(dataset.data, columns=dataset.feature_names)
```

```python
#adding target column to the data frame
data_frame['label'] = dataset.target
```

separating features and target

```python
X = data_frame.drop(columns='label', axis=1)
Y =data_frame['label']
```

```python
print(X)
```

```python
print(Y)
```

splitting data into training data and testing data

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2,random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

standardizing the data

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_train_std = scaler.fit_transform(X_train)
X_test_std = scaler.transform(X_test)
```

```python
print(X_train_std)
```

**BUILDING NEURAL NETWORK**

```python
#importing tensorflow and keras
import tensorflow as tf
tf.random.set_seed(3) #to get the same accuracy score each time u run nn
from tensorflow import keras #keras is a wrapper for handling backend
```

```python
#setting up layers of nn
#no. of neurons = no. of features in ip layer
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)), #ip layer by converting matrix ip like xtrain ytrain to single dimensional array; 30 neurons
    keras.layers.Dense(20, activation='relu'), #hidden layer; 20 =no. of neurons u want (op accuracy will depend on no. of neurons)
    keras.layers.Dense(2, activation='sigmoid') #op layer; 2 neurons bcuz we have 2 classes
])
```

```python
#Compiling the neural network
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
history = model.fit(X_train_std, y_train, validation_split=0.1, epochs=10)
```

visualizing accuracy and loss

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['training data', 'validation data'], loc='lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['training data', 'validation data'], loc='upper right')
```

accuracy of model on test data

```python
loss, accuracy = model.evaluate(X_test_std, y_test)
print(accuracy)
```

building the predictive system

```python
input_data = (20.57, 17.77, 132.9, 1326.0, 0.08474, 0.07864, 0.0869, 0.07017, 0.1812, 0.05667, 0.5435, 0.7339, 3.398, 74.08, 0.005225, 0.01308, 0.0186, 0.0134, 0.01389, 0.003532, 24.99, 23.41, 158.8, 1956.0, 0.1238, 0.1866, 0.2416, 0.186, 0.275, 0.08902)

#change the input data to numpy array
input_data_as_numpy_array = np.asarray(input_data)

#reshape the numpy array as we are predicting for one data point
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

#standardizing the ip data
input_data_std = scaler.transform(input_data_reshaped)

prediction = model.predict(input_data_std)
print(prediction)
prediction_label = [np.argmax(prediction)]
print(prediction_label)

if(prediction_label[0] == 0):
  print('the tumor is malignant')
else:
  print('the tumor is benign')
```
